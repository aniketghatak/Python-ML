```python
import numpy as np
import matplotlib.pyplot as plt
```

###  A normal distribution graph is also known as the bell curve because of it's characteristic shape of a bell.

```python
data = np.random.normal(50, 10, 100000) ## We specify that the mean value is 50, and the standard deviation is 10.
plt.hist(data,100) ## draw a histogram with 100 bars
plt.show()
```

```python
data_2 = np.random.normal(15, 5, 100000) ## We specify that the mean value is 15, and the standard deviation is 5.
plt.hist(data_2,50) ## draw a histogram with 50 bars
plt.show()
```
