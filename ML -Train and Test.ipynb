```python
import numpy
import matplotlib.pyplot as plt
numpy.random.seed(2)
```

```python
x = numpy.random.normal(3, 1, 150)
y = numpy.random.normal(150, 40, 150) / x

plt.scatter(x, y)
plt.show()
```

### Training set is random selection of 80% of the original data. The testing set is remaining 20%

```python
train_x = x[:80]
train_y = y[:80]

test_x = x[80:]
test_y = y[80:]
```

### Display the training set

```python
plt.scatter(train_x, train_y)
plt.show()
```

### Display the testing set

```python
plt.scatter(test_x, test_y)
plt.show()
```

### Polynomial Regression

```python
mymodel = numpy.poly1d(numpy.polyfit(train_x, train_y, 4))

myline = numpy.linspace(0, 6, 100)

plt.scatter(train_x, train_y)
plt.plot(myline, mymodel(myline))
plt.show()
```

### R2- Training Set

```python
from sklearn.metrics import r2_score

r2 = r2_score(train_y, mymodel(train_x))

print(r2)
```

### R2 Testing Set

```python
r2 = r2_score(test_y, mymodel(test_x))
print(r2)
```
