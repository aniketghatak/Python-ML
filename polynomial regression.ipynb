```python
import matplotlib.pyplot as plt
```

```python
score=[100,90,80,68,66,69,68,65,70,70,75,76,78,79,60,79,69,78]
age=[21,22,23,25,26,27,28,29,30,22,23,24,25,26,28,29,21,22]
plt.scatter(age, score)
plt.show()
```

```python
import numpy as np
```

```python
mymodel = np.poly1d(np.polyfit(age, score, 3))

myline = np.linspace(20, 30, 100)

plt.scatter(age, score)
plt.plot(myline, mymodel(myline))
plt.show()
```

```python
from sklearn.metrics import r2_score
```

```python
mymodel = np.poly1d(np.polyfit(age, score, 3))
print(r2_score(score, mymodel(age)))
```
